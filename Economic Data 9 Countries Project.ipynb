#### Let's look at the inflation rate data of India from 1980 - 2020

```python
df_india = df[df['country'] == 'India']
df_india.head()
```

#### Plot the inflation rate of India

```python
plt.figure(figsize = (15,6))
sns.set_style("whitegrid")
plt.title("India Inflation Rate 1980-2020", fontsize=20, color = 'green')
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Inflation rate", fontsize = 15)
sns.lineplot(x = 'year', y = 'inflationrate', data = df_india , color = 'green')
plt.show()
```

##### PLot Per Capita Income of India 1980-2020 period

```python
plt.figure(figsize = (15,6))
sns.set_style("whitegrid")
plt.title("India Per Capita Income 1980-2020", fontsize=20, color = 'green')
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Per Capita Income", fontsize = 15)
sns.lineplot(x = 'year', y = 'percapitaincome', data = df_india , color = 'green')
plt.show()
```

#### Plot Unemployment Rate in India from 1980-2020

```python
plt.figure(figsize = (15,6))
plt.title("India Unemployment Rate 1980 - 2020", fontsize = 20, color = 'green')
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Unemployment Rate", fontsize = 15)
sns.lineplot(x = 'year', y = 'unemploymentrate', data = df_india)
plt.show()
```

#### PLot India's Manufacturing Output 1980 - 2020

```python
plt.figure(figsize = (15,6))
plt.title("India Manufacturing Output 1980 - 2020", fontsize = 20, color = 'brown')
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Manufacturing Output", fontsize = 15)
sns.lineplot(x = 'year', y = 'manufacturingoutput', data = df_india, color = 'green')
plt.show()
```

#### Plot GDP Percentage of India 1980 - 2020

```python
plt.figure(figsize = (15,6))
plt.title("India Gross Domestic Product Rate 1980 - 2020", fontsize = 20, color = 'brown')
sns.set_style("darkgrid")
plt.xlabel("Year", fontsize = 15)
plt.ylabel("Manufacturing Output", fontsize = 15)
sns.lineplot(x = 'year', y = 'gdppercent', data = df_india, color = 'green')
plt.show()
```

#### FInd top five (5) countries in terms of Per Capita Income & Plot a bargraph on it

```python
top_5_pci = df['percapitaincome'].groupby(df['country']).mean().sort_values(ascending=False).reset_index()[:5]
top_5_pci
```

```python
plt.figure(figsize = (8,6))
plt.title("Top Five (5) Countries wrt Per Capita Income", fontsize = 18, color = 'brown')
plt.xlabel("Country", fontsize = 14.5, color = 'green')
plt.ylabel("Per Capita Income", fontsize = 14.5, color = 'green')
plt.xticks(rotation = 60)
sns.barplot(x = 'country', y = 'percapitaincome', data = top_5_pci)
plt.show()
```

#### Rank Countries in increasing order of unemployment rate

```python
df_urt = df['unemploymentrate'].groupby(df['country']).mean().sort_values().reset_index()
df_urt
```

```python
plt.figure(figsize = (10,5))
plt.title("Countries with increasing order of their Unemployment rate", fontsize = 18, color = 'brown')
plt.xlabel("Country", fontsize = 14.5, color = 'green')
plt.ylabel("Unemployment Rate", fontsize = 14.5, color = 'green')
plt.xticks(rotation = 60)
sns.barplot(x = 'country', y = 'unemploymentrate', data = df_urt)
plt.show()
```

#### Find top 5 countries that have faced high inflation rates during 1980 -2020

```python
top_5_ir = df['inflationrate'].groupby(df['country']).mean().sort_values(ascending=False).reset_index()[:5]
top_5_ir
```

```python
plt.figure(figsize = (7,4.6))
plt.title("Top Five (5) Countries with High Inflation Rate", fontsize = 18, color = 'brown')
plt.xlabel("Country", fontsize = 14.5, color = 'green')
plt.ylabel("Inflation Rate", fontsize = 14.5, color = 'green')
plt.xticks(rotation = 60)
sns.barplot(x = 'country', y = 'inflationrate', data = top_5_ir, palette="magma")
plt.show()
```

#### Arrange Countries in descending order wrt GDP Percentage

```python
gdp_coun = df['gdppercent'].groupby(df['country']).mean().sort_values(ascending=True).reset_index()
gdp_coun
```

```python
plt.figure(figsize = (10,5))
plt.title("Countries with increasing order of their GDP growth", fontsize = 18, color = 'brown')
plt.xlabel("Country", fontsize = 14.5, color = 'green')
plt.ylabel("GDP Rate", fontsize = 14.5, color = 'green')
plt.xticks(rotation = 60)
sns.barplot(x = 'country', y = 'gdppercent', data = gdp_coun, palette="rocket")
plt.show()
```
